```python
import pandas as pd
import statsmodels.formula.api as smf
import warnings
warnings.filterwarnings('ignore')

raw_t8 = pd.read_excel('data/raw/chetty2014_online_data_tables.xls',
                       sheet_name='Online Data Table 8', header=6)
raw_t8['CZ'] = pd.to_numeric(raw_t8['CZ'], errors='coerce')
raw_t8 = raw_t8.dropna(subset=['CZ'])

county = pd.read_csv('data/raw/county_outcomes_simple.csv')
county['cz'] = pd.to_numeric(county['cz'], errors='coerce')

KEEP = ['CZ','Racial Segregation','Gini Bottom 99%',
        'Test Score Percentile (Income Adjusted',
        'Social Capital Index',
        'Fraction of Children with Single Mothers',
        'Frac. with Commute < 15 Mins']
merged = county.merge(raw_t8[KEEP], left_on='cz', right_on='CZ', how='inner')
merged = merged.rename(columns={
    'Racial Segregation':                       'racial_seg',
    'Gini Bottom 99%':                          'gini99',
    'Test Score Percentile (Income Adjusted':   'school_quality',
    'Social Capital Index':                     'social_capital',
    'Fraction of Children with Single Mothers': 'single_mothers',
    'Frac. with Commute < 15 Mins':            'commute_lt15'
})

COLS = ['kfr_pooled_pooled_p25','racial_seg','gini99','school_quality',
        'social_capital','single_mothers','commute_lt15']
for col in COLS:
    merged[col] = pd.to_numeric(merged[col], errors='coerce')

df = merged[COLS].dropna()
print(f"N = {len(df)}")

model = smf.ols(
    'kfr_pooled_pooled_p25 ~ racial_seg + gini99 + school_quality + '
    'social_capital + single_mothers + commute_lt15',
    data=df
).fit(cov_type='HC3')
print(model.summary())
```
