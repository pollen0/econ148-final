```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import warnings
warnings.filterwarnings('ignore')

# Load Table 8 with correct header row
t8 = pd.read_excel('data/raw/chetty2014_online_data_tables.xls',
                   sheet_name='Online Data Table 8', header=6)
t8 = t8.rename(columns={
    'CZ': 'cz',
    'Racial Segregation': 'racial_seg',
    'Gini Bottom 99%': 'gini99',
    'Test Score Percentile (Income Adjusted': 'school_quality',
    'Social Capital Index': 'social_capital',
    'Fraction of Children with Single Mothers': 'single_mothers',
    'Frac. with Commute < 15 Mins': 'commute_lt15'
})
t8['cz'] = pd.to_numeric(t8['cz'], errors='coerce')
t8 = t8.dropna(subset=['cz'])

# Load crosswalk
cw = pd.read_csv('data/raw/cty_cz_st_crosswalk.csv')
print("Crosswalk columns:", cw.columns.tolist())
print("Crosswalk sample:", cw.head(2))
```

```python
# county already has 'cz' — merge t8 covariates directly
KEEP = ['cz', 'racial_seg', 'gini99', 'school_quality',
        'social_capital', 'single_mothers', 'commute_lt15']

merged = county.merge(t8[KEEP], on='cz', how='inner')

print("merged shape:", merged.shape)
print("merged columns sample:", merged[['state', 'county', 'cz',
      'kfr_pooled_pooled_p25', 'racial_seg']].head(3))
```

```python
OUTCOME = 'kfr_pooled_pooled_p25'
FEATURES = ['racial_seg', 'gini99', 'school_quality',
            'social_capital', 'single_mothers', 'commute_lt15']

df = merged[FEATURES + [OUTCOME]].dropna()
print(f"Analytic sample: {len(df)} counties")

formula = f"{OUTCOME} ~ " + " + ".join(FEATURES)
model = smf.ols(formula, data=df).fit(cov_type='HC3')
print(model.summary())
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import warnings
warnings.filterwarnings('ignore')

# Load Table 8 with correct header row
raw_t8 = pd.read_excel('data/raw/chetty2014_online_data_tables.xls',
                       sheet_name='Online Data Table 8', header=6)

t8 = raw_t8.rename(columns={
    'CZ':                                        'cz',
    'Racial Segregation':                        'racial_seg',
    'Gini Bottom 99%':                           'gini99',
    'Test Score Percentile (Income Adjusted':    'school_quality',
    'Social Capital Index':                      'social_capital',
    'Fraction of Children with Single Mothers':  'single_mothers',
    'Frac. with Commute < 15 Mins':             'commute_lt15'
})
t8['cz'] = pd.to_numeric(t8['cz'], errors='coerce')
t8 = t8.dropna(subset=['cz'])

# Load county outcomes — already has 'cz' column built in
county = pd.read_csv('data/raw/county_outcomes_simple.csv')

# Merge directly on cz — no crosswalk needed
KEEP = ['cz', 'racial_seg', 'gini99', 'school_quality',
        'social_capital', 'single_mothers', 'commute_lt15']
merged = county.merge(t8[KEEP], on='cz', how='inner')

# Build analytic sample
OUTCOME  = 'kfr_pooled_pooled_p25'
FEATURES = ['racial_seg', 'gini99', 'school_quality',
            'social_capital', 'single_mothers', 'commute_lt15']
df = merged[[OUTCOME] + FEATURES].dropna()
print(f"Analytic sample: {len(df)} counties\n")

# OLS with HC3 robust standard errors — sm.OLS is faster than smf.ols
X = sm.add_constant(df[FEATURES].astype(float))
y_ols = df[OUTCOME].astype(float)
model = sm.OLS(y_ols, X).fit(cov_type='HC3')
print(model.summary())
```
